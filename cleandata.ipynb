```python
import pandas as pd
```

```python
total_data= pd.read_csv("Wimbledon_featured_matches.csv")
```

```python
Carlos_player1= total_data.loc[total_data["player1"]=="Carlos Alcaraz"]
Carlos_player2= total_data.loc[total_data["player2"]=="Carlos Alcaraz"]
Carlos_Alcaraz = pd.concat([Carlos_player1,Carlos_player2],ignore_index=True)
#it seeems Carlos only appear as the first player
```

```python
Novac_player1= total_data.loc[total_data["player1"]=="Novak Djokovic"]
Novac_player2= total_data.loc[total_data["player2"]=="Novak Djokovic"]
Novak_Djokovic = pd.concat([Novac_player1,Novac_player2],ignore_index=True)
Novak_Djokovic
#it seeems Carlos only appear as the first player
```
